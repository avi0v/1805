```python
#import 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#loading data
uni=pd.read_csv('Universities.csv')
uni
```

```python
#data cleaning AND data scaling 
uni.info()
```

```python
#use normalize scale
def ns(a):
    b=(a-a.mean())/a.std()
    return b
```

```python
#scaling 
x=ns(uni.iloc[:,1:])
x
```

```python
'''DENDROGRAM'''
```

```python
import scipy.cluster.hierarchy as sch 
dendro=sch.dendrogram(sch.linkage(x,method='complete'))
```

```python
'''MAKING MODEL'''
```

```python
#loading model and then fitting feature that are being used in clustering /grouping

from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
hc
```

```python
#fit data in the model 
# in above we load data in a variable hc
y=hc.fit_predict(x)
```

```python
y
```

```python
uni['clusterid']=y
```
